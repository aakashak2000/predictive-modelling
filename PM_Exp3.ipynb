```python
#treating outliers
#creating a mask for upper and lower bound beyond which the observations are considered outliers
#outliers are replaced with values of the lower and upper bounds accodingly
for col in df.columns:
    print("capping the ",col)
    if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
        df_mean = df[col].mean()
        df_std = df[col].std()
        cut_off = df_std * 3
        lower, upper = df_mean - cut_off, df_mean + cut_off
        df[col][df[col] < lower] = lower
        df[col][df[col] > upper] = upper
    else:
        df[col]=df[col]
```
